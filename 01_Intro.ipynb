### Definición de funciones

Las funciones se definen con la palabra clave `def`

```python
def funcion(argumento1, argumento2, argumento_con_default=None, 
            *argumentos_posicionales_aninimos, **otros_argumentos_con_nombre)
    # hacer algo
    algo = 0
    return algo
```

````python
def func1(a, b=1, *args, **kwargs):
    print("a =", a)
    print("b =", b)
    for i, arg in enumerate(args):
        print("elemento {0} de args es igual a {1}".format(i, arg))
    for i, arg in enumerate(kwargs):
        print("elemento {0} de kwargs es igual a {1}".format(arg, kwargs[arg]))
````

### Definición de Clases

#### Generadores

Los generadores son un tipo especial de iteradores. A diferencia de una lista, por ejemplo, no contienen en memoria todos sus elementos, sino que los van "generando" a medida que le son requeridos

````python
generator = (a for a in range(10))
generator
````

````python
for a in generator:
    print(a)
````

````python
generator
````

````python
list(generator)
````

Los objetos range son generadores, también map, zip, enumerate y una gran cantidad de funciones integradas (de python 3 en adelante por lo general)
La forma de hacer un generador a medida, es usando la palabra clave "yield"

````python
def gen():
    while True:
        yield np.random.rand()

g = gen()
print(next(g))
````

````python
for rn in g:
    print(rn)
    if rn > 0.8: 
        break
    else:
        continue
    print("esto no corre")
````

Los generadores dejan de entregar valores cuando se tira una excepción de fin de iteración

````python
def gen():
    while True:
        rn = np.random.rand()
        yield rn 
        if rn > 0.8: break
            
def gen_2():
    while True:
        rn = np.random.rand()
        yield rn 
        if rn > 0.8: raise StopIteration
````

````python
for rn in gen():
    print(rn)

print("*"*10)
for rn in gen_2():
    print(rn)
````
